```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
%matplotlib notebook
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
h = 1
sd = 1
n = 50

def gen_data(n, h, sd1, sd2):
    # mean, sd, number of points to generate
    x1 = ss.norm.rvs(-h, sd1, n)
    y1 = ss.norm.rvs(0, sd1, n)

    x2 = ss.norm.rvs(h, sd2, n)
    y2 = ss.norm.rvs(0, sd2, n)
    return (x1, y1, x2, y2)
```

```python
#(x1, y1, x2, y2) = gen_data(n, h, sd1, sd2)
(x1, y1, x2, y2) = gen_data(n, h, 1, 1.5)
(x1, y1, x2, y2) = gen_data(1000, 1.5, 1, 1.5)
```

```python
# no need to return anything in this function
def plot_data(x1, y1, x2, y2):
    %matplotlib inline
    plt.figure()
    #plt.figure(figsize=(12,7))
    plt.plot(x1, y1, 'o', ms=2) # blue
    plt.plot(x2, y2, 'o', ms=2) # orange broader cloud because greater sd (1.5 > 1)
    plt.xlabel("$X_1$")
    plt.ylabel("$X_2$")
    
plot_data(x1, y1, x2, y2)
```
