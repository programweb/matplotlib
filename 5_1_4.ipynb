```python
# least squares.
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt

# also see:
#    https://github.com/dnackat/python-for-research/blob/master/LinearRegression/linearRegression.py
# it is in 5.1.2_quiz.py
n = 100
beta_0 = 5
beta_1 = 2
np.random.seed(1)
# 'n' random variables that are distributed between 0 and 1
x = 10 * ss.uniform.rvs(size=n)
# Add random noise (gaussian) with ss.norm (mean=0, n=100)
y = beta_0 + beta_1 * x + ss.norm.rvs(loc=0, scale = 1, size = n)

rss = [] # residual sum of squares
# quiz only do this: slopes = np.arange(-10, 15, 0.001)
slopes = np.arange(-10, 15, 0.01)
for slope in slopes:
    rss.append(np.sum((y - beta_0 - slope * x)**2))

# print(rss)
```

```python
ind_min = np.argmin(rss)
print(ind_min)
print("Instructor's estimate for slope: 2.0")
print("My estimate for the slope: ", slopes[ind_min])
```

```python
plt.figure()
plt.plot(slopes, rss)
plt.xlabel("Slope")
plt.ylabel("RSS")
```
